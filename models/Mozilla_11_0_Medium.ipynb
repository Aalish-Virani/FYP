```python
# Install required libraries
!pip install -q transformers torchaudio librosa
!apt install -y ffmpeg  # Ensure ffmpeg is installed

import torch
from transformers import pipeline
from IPython.display import Audio

output_wav = "converted_audio.wav"

# Verify the file exists
!ls -lh "{input_mp3}"

# Play the audio to verify (optional)
Audio(input_mp3)

# Load the pipeline
pipe = pipeline("automatic-speech-recognition",
                model="ihanif/whisper-medium-urdu",
                device="cuda" if torch.cuda.is_available() else "cpu")

# Convert MP3 to WAV (16kHz mono)
!ffmpeg -i "{input_mp3}" -ar 16000 -ac 1 "{output_wav}"

# Verify WAV file was created
!ls -lh "{output_wav}"

# Transcribe the audio
try:
    transcription = pipe(output_wav)
    print("\nTranscription:")
    print(transcription["text"])

    # Save transcription to file
    with open("transcription.txt", "w", encoding="utf-8") as f:
        f.write(transcription["text"])
    print("\nTranscription saved to transcription.txt")

except Exception as e:
    print(f"Error during transcription: {e}")
    print("Make sure the audio file is in the correct format (16kHz WAV)")
```
