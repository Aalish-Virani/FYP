```python

# ðŸ“¦ Step 2: Import required libraries
import whisper
import os

# ðŸ“‚ Step 3: Set file names
output_wav = "converted_audio.wav"

# ðŸŽµ Step 4: Convert MP3 to WAV (mono, 16kHz)
!ffmpeg -i "{input_mp3}" -ar 16000 -ac 1 "{output_wav}" -y

# ðŸ§  Step 5: Load the Whisper model
model = whisper.load_model("small")  # or use "medium" if you're low on RAM

# ðŸ“ Step 6: Transcribe with word timestamps
result = model.transcribe(output_wav, language='ur', word_timestamps=True, verbose=False)

# ðŸ“œ Step 7: Print full transcript first
full_text = result["text"]
print("ðŸ—£ï¸ Full Urdu Transcript:\n")
print(full_text)

# ðŸ§© Step 8: Print word-level timestamps and probabilities
print("\nðŸ” Word-by-Word Breakdown with Timestamps and Confidence:\n")
for segment in result['segments']:
    for word in segment['words']:
        word_text = word['word'].strip()
        start = word['start']
        end = word['end']
        confidence = word['probability']
        print(f"{word_text} | â± {start:.2f} - {end:.2f} | ðŸ“Š Confidence: {confidence:.2f}")

# ðŸ’¾ Step 9: Optionally save the output to text file
with open("urdu_transcript_detailed.txt", "w", encoding="utf-8") as f:
    f.write("Full Transcript:\n")
    f.write(full_text + "\n\n")
    f.write("Word-level Details:\n")
    for segment in result['segments']:
        for word in segment['words']:
            line = f"{word['word'].strip()} | {word['start']:.2f} - {word['end']:.2f} | Confidence: {word['probability']:.2f}\n"
            f.write(line)

print("\nâœ… Transcription and details saved to urdu_transcript_detailed.txt")
```
