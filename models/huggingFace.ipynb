```python
import pandas as pd
from datasets import Dataset, Audio

# Load your Kaggle dataset
df = pd.read_csv('/path/to/your/transcriptions.csv')  # Update path
df['audio_path'] = '/path/to/audio/files/' + df['audio_filename']  # Update path

# Convert to HuggingFace Dataset
dataset = Dataset.from_pandas(df)

# Cast the audio column to Audio feature
dataset = dataset.cast_column("audio_path", Audio())

# Rename columns to match Whisper's expected format
dataset = dataset.rename_column("transcription_text", "sentence")
dataset = dataset.rename_column("audio_path", "audio")

# Split dataset into train and test
dataset = dataset.train_test_split(test_size=0.1)
```
